```python
import numpy as np
import open3d as o3d
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
pcd = o3d.io.read_point_cloud("TestData/test_face.pcd")
points = np.array(pcd.points)

X = points[:, 0]
Y = points[:, 1]
Z = points[:, 2]

fig = plt.figure()
ax = Axes3D(fig)

#軸にラベルを付けたいときは書く
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

#.plotで描画
#linestyle='None'にしないと初期値では線が引かれるが、3次元の散布図だと大抵ジャマになる
#markerは無難に丸
ax.plot(X,Y,Z,marker="o",linestyle='None', markersize=1)

#最後に.show()を書いてグラフ表示
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# 真正面
axes[0].plot(points[:, 0], points[:, 1], 'o', markersize=4)
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")
# 真横
axes[1].plot(points[:, 2], points[:, 1], 'o', markersize=4)
axes[1].set_xlabel("Z")
axes[1].set_ylabel("Y")
# 真上
axes[2].plot(points[:, 0], points[:, 2], 'o', markersize=4)
axes[2].set_xlabel("X")
axes[2].set_ylabel("Z")
# plotしたデータごとに名前をつけてる
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

for idx, ax in zip([0, 1, 2], axes):
    deleted_points = np.delete(points, idx, 1)
    ax.plot(deleted_points[:, 0], deleted_points[:, 1], 'o', markersize=4)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

fig, axes = plt.subplots(1, 3, figsize=(20, 6))

for idx, ax in zip([0, 1, 2], axes):
    deleted_points = np.delete(points, idx, 1)
    scaled_points = scaler.fit_transform(deleted_points)
    ax.plot(scaled_points[:, 0], scaled_points[:, 1], 'o', markersize=4)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

fig, axes = plt.subplots(1, 3, figsize=(20, 6))

for idx, ax in zip([0, 1, 2], axes):
    deleted_points = np.delete(points, idx, 1)
    scaled_points = scaler.fit_transform(deleted_points)
    ax.plot(scaled_points[:, 0], scaled_points[:, 1], 'o', markersize=4)
```
